### Clean Up Relevant Columns

```python
cols = ['IMO Number', 'Name', 'Technical efficiency', 'Ship type', 'DoC issue date', 'DoC expiry date',
       'Total fuel consumption [m tonnes]', 'Annual Time spent at sea [hours]',
       'Annual average CO₂ emissions per distance [kg CO₂ / n mile]',
       'Annual average CO₂ emissions per transport work (mass) [g CO₂ / m tonnes · n miles]',
       'Verifier Name']
df = df[cols]
```

```python
df['technical_efficiency_number'] = df['Technical efficiency'].str.extract(r'(\d*[\.]\d*)')
df['rating_type'] = df['Technical efficiency'].str.extract(r'^(\w*\s)')
df['rating_type'] = df['rating_type'].str.strip()

df.head()
```

```python
# Rename columns
col_names = {'IMO Number': 'imo', 'Name':'ship_name', 'Technical efficiency':'technical_efficiency',
             'Ship type': 'ship_type', 'DoC issue date': 'issue_date', 'DoC expiry date':'expiry_date',
             'Total fuel consumption [m tonnes]': 'fuel_consumption', 'Annual Time spent at sea [hours]':'sea_time',
             'Annual average CO₂ emissions per distance [kg CO₂ / n mile]': 'co2_distance',
             'Annual average CO₂ emissions per transport work (mass) [g CO₂ / m tonnes · n miles]':'co2_transport',
            'Verifier Name':'verifier'}

df.rename(columns = col_names, inplace = True)

df.head()
```

```python
# cleanup EIV and EEDI
import numpy as np

df['EIV'] = np.where(df['rating_type'] =='EIV', df['technical_efficiency_number'],np.nan)
df['EEDI'] = np.where(df['rating_type'] =='EEDI', df['technical_efficiency_number'],np.nan)

df.head()
```

```python
# Clean out cols that are useless now

df.drop(['technical_efficiency_number', 'rating_type','technical_efficiency'], axis = 1, inplace = True)

df.head()
```

```python
# Change date formats
df['issue_date']=pd.to_datetime(df['issue_date']).dt.date
df['expiry_date']=pd.to_datetime(df['expiry_date']).dt.date
```

```python
# Cleanup some weird values
df = df.replace('Division by zero!', np.nan)
```

```python
# Drop columns where issue date or expiry data are null
df.dropna(subset = ['issue_date','expiry_date', 'verifier','EEDI'],inplace = True)
```

```python
# add in ship and verifier dim surrogate key

df_v = pd.read_csv('verifier_names_to_key.csv')
df_s = pd.read_csv('ship_dimension.csv')

df = df.merge(df_v, left_on = 'verifier', right_on = 'Verifier Name').drop(columns= ['Verifier Name'])
df.head()
```

```python
df = df.merge(df_s[['imo', 'ship']], on = 'imo')
df.head()
```

```python
# add in date dim surrogate key
df['issue_date_key'] = df['issue_date'].apply(lambda x: x.strftime('%Y%m%d'))
df['expiry_date_key'] = df['expiry_date'].apply(lambda x: x.strftime('%Y%m%d'))

df.head()
```

```python
# rename key columns
col_names = {'Verifier Key':'verifier_key', 'ship': 'ship_key'}
df.rename(columns = col_names, inplace = True)

df.head()
```

```python
# drop columns that are now captured in the dim table
# leave the issue_date and imo there
# drop EIV also as it seems to be empty
df.drop(['verifier', 'EIV', 'issue_date','expiry_date','ship_name','ship_type'], axis = 1, inplace = True)
```
